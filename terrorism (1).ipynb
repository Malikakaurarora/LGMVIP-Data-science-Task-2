# Terrorist Activities each Year

```python
year=data['iyear'].unique()
years_count= data['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize=(25,20)) 
sns.barplot(x=year, y=years_count, palette='rocket') 
plt.xticks(rotation=40)
plt.title("Attacks", fontsize=22)
plt.show()
```

# Top terrorist group attack

```python
data['gname'].value_counts().to_frame().drop("Unknown").head(30).plot(kind="bar",color="green", figsize=(30,20)) 
plt.title("Top terrorist group attack", fontsize=25)

plt.show()
```

# Terrorist Activities by Region in each Year through Area Plot

```python
pd.crosstab(data.iyear, data.region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```
